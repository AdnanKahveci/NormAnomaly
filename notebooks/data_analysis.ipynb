```python
import pandas as pd # type: ignore
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore
import seaborn as sns # type: ignore

# Geriye dönük uyarı
import warnings
warnings.filterwarnings('ignore')
```

```python
# Dosya yolunu belirtin
file_path = 'C:/Users/kahve/Desktop/AnomaFlow/data/raw/creditcard.csv'

# Veriyi yükleyin
df = pd.read_csv(file_path)

# Veri hakkında genel bilgi
print(df.info())
print(df.describe())
```

```python
# Sınıf dağılımını inceleyin
sns.countplot(x='Class', data=df)
plt.title('Sınıf Dağılımı')
plt.show()

# Özniteliklerin dağılımını inceleyin
df[['Amount', 'Time']].hist(bins=50, figsize=(10, 5))
plt.show()
```

```python
# Eksik verileri kontrol edin
print(df.isnull().sum())

# Anomalileri kontrol edin
print(df[df['Amount'] > 1000].head())
```

```python
# Korelasyon matrisini hesaplayın
corr_matrix = df.corr()

# Korelasyon matrisini görselleştirin
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Korelasyon Matrisi')
plt.show()
```
